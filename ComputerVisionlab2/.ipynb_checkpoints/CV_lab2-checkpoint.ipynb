```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('carkyo.jpg')
plt.imshow(img[:,:,::-1])
ing_l,img_w,img_h=img.shape
```

```python
logo = cv2.imread('tekomoro.png')
logo_l,logo_w,logo_h=logo.shape
scale_percent = 150/logo_w
width = int(logo.shape[1] * scale_percent )
height = int(logo.shape[0] * scale_percent)
print(width,height)
logo = cv2.resize(logo,(width, height))
plt.imshow(logo[:,:,::-1])
```

```python
roi = img[img_h-34:img_h-10,img_w-160:img_w-10]
print(roi.shape[1],roi.shape[0])
plt.imshow(roi[:,:,::-1])
```

```python
gray = cv2.cvtColor(logo, cv2.COLOR_BGR2GRAY) 
plt.imshow(gray, cmap='gray');
```

```python
ret, mask = cv2.threshold(gray, 50, 255, cv2.THRESH_BINARY_INV)
plt.imshow(mask, cmap='gray');
```

```python
mask_inv = cv2.bitwise_not(mask)
plt.imshow(mask_inv, cmap='gray');
```

```python
img_fg = cv2.bitwise_and(logo, logo, mask=mask_inv)
plt.imshow(img_fg[:,:,::-1]);
```

```python
roi = cv2.bitwise_and(roi, roi, mask=mask)
plt.imshow(roi[:,:,::-1]);
```

```python
logo = cv2.add(roi, img_fg)
plt.imshow(logo[:,:,::-1])
```

```python
copy = img.copy()
copy[img_h-34:img_h-10,img_w-160:img_w-10]=logo
plt.imshow(copy[:,:,::-1]);
```

```python
cv2.imwrite("out.png",copy)
```
