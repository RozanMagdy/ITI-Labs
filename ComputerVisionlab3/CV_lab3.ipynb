```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

## Task 1

```python
img = cv2.imread('carkyo.jpg')
r = img[:,:,0]
g = img[:,:,1]
b = img[:,:,2]
f, (ax1, ax2, ax3, ax4) = plt.subplots(1, 4, figsize=(25,20))
ax1.set_title('Original Image')
ax1.imshow(img)
ax2.set_title('R channel')
ax2.imshow(r, cmap='Reds')
ax3.set_title('g channel')
ax3.imshow(g, cmap='Greens')
ax4.set_title('b channel')
ax4.imshow(b, cmap='Blues');
#ing_l,img_w,img_h=img.shape
```

```python
hsv = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)
h = hsv[:,:,0]
s = hsv[:,:,1]
v = hsv[:,:,2]
f, (ax1, ax2, ax3, ax4) = plt.subplots(1, 4, figsize=(25,20))
ax1.set_title('Original Image')
ax1.imshow(hsv)
ax2.set_title('H channel')
ax2.imshow(h, cmap='hsv')
ax3.set_title('S channel')
ax3.imshow(s, cmap='hsv')
ax4.set_title('V channel')
ax4.imshow(v, cmap='hsv');
```

```python
h = hsv[:,:,0]-5
s = hsv[:,:,1]+10
v = hsv[:,:,2]*1.25
h[h>255] = 255
h[h<0] = 0
s[s>255] = 255
s[s<0] = 0
v[v>255] = 255
v[v<0] = 0
f, (ax1, ax2, ax3, ax4) = plt.subplots(1, 4, figsize=(25,20))
ax1.set_title('Original Image')
ax1.imshow(hsv)
ax2.set_title('H channel')
ax2.imshow(h, cmap='hsv')
ax3.set_title('S channel')
ax3.imshow(s, cmap='hsv')
ax4.set_title('V channel')
ax4.imshow(v, cmap='hsv');
```

```python
#hsv_image= cv2.merge(np.array([h, s, v]).astype(np.uint8))
out_task1 = cv2.cvtColor(hsv, cv2.COLOR_HSV2BGR)
plt.imshow(out_task1[:,:,::-1])
cv2.imwrite('out_task1.png',out_task1)
```

## Task 2

```python
yCrCb = cv2.cvtColor(img, cv2.COLOR_RGB2YCrCb)
y = yCrCb[:,:,0]
cr = yCrCb[:,:,1]
cb = yCrCb[:,:,2]

f, (ax1, ax2, ax3, ax4) = plt.subplots(1, 4, figsize=(25,20))
ax1.set_title('Original Image')
ax1.imshow(img)
ax2.set_title('Y channel')
ax2.imshow(y, cmap='PiYG_r')
ax3.set_title('Cr channel')
ax3.imshow(cr, cmap='PiYG_r')
ax4.set_title('Cb channel')
ax4.imshow(cb, cmap='PiYG_r');
```

```python
yCrCb = cv2.cvtColor(img, cv2.COLOR_RGB2YCrCb)
y = yCrCb[:,:,0]*1.25
cr = yCrCb[:,:,1]
cb = yCrCb[:,:,2]
y[y>255] = 255
y[y<0] = 0
f, (ax1, ax2, ax3, ax4) = plt.subplots(1, 4, figsize=(25,20))
ax1.set_title('Original Image')
ax1.imshow(img)
ax2.set_title('Y channel')
ax2.imshow(y, cmap='PiYG_r')
ax3.set_title('Cr channel')
ax3.imshow(cr, cmap='PiYG_r')
ax4.set_title('Cb channel')
ax4.imshow(cb, cmap='PiYG_r');
```

```python
#yCrCb_image= cv2.merge(np.array([y, cr, cb]).astype(np.uint8))
out_task2 = cv2.cvtColor(yCrCb, cv2.COLOR_YCrCb2BGR)
plt.imshow(out_task2[:,:,::-1])
cv2.imwrite('out_task2.png',out_task2)
```
